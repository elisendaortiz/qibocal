```python
import os

from qibocal import create_calibration_platform
from qibocal.protocols import rabi_amplitude
```

```python
# Over-writing / Expliciting the path to the qubit local platform since
# 'QIBOLAB_PLATFORMS' typically (global laptop setup) points to platforms repo
# QIBOLAB_PLATFORMS=/Users/elis/Desktop/PROJECTS/Quantum/qibocal/platforms
os.environ["QIBOLAB_PLATFORMS"] = (
    "/Users/elis/Desktop/PROJECTS/Quantum/qibocal/platforms"
)
print(f"QIBOLAB_PLATFORMS: {os.environ['QIBOLAB_PLATFORMS']}")
```

```python
platform = create_calibration_platform("qubit")
```

```python
# define qubits where the protocols will be executed
targets = [0]

# define protocol parameters
params = rabi_amplitude.parameters_type.load(
    dict(
        min_amp=0.01,
        max_amp=0.2,
        step_amp=0.02,
        nshots=2000,
        pulse_length=40,
    )
)

# acquire
platform.connect()
data, acquisition_time = rabi_amplitude.acquisition(
    params=params, platform=platform, targets=targets
)


platform.disconnect()
```

```python
# post-processing
results, fit_time = rabi_amplitude.fit(data=data)
```

```python
# To check the problem with results --> fit in function rabi_amplitude
# import inspect
# print(inspect.signature(rabi_amplitude.report))
```

```python
# visualize the results
plots, table = rabi_amplitude.report(data=data, fit=results, target=targets[0])
plots[0].show(renderer="browser")

# save to file
# create directory if it doesn't exist
os.makedirs("./getting_started/plots", exist_ok=True)
plots[0].write_html("./getting_started/plots/rabi_amplitude_plot.html")
```
